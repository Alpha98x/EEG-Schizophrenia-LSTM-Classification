```python
#
import os
import numpy as np
import pandas as pd
import mne
from scipy.signal import stft
from torch.utils.data import Dataset, DataLoader
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, f1_score, roc_curve, auc, confusion_matrix
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import torch.nn.functional as F

class EEGDataset(Dataset):
    def __init__(self, eeg_data, panss_scores):
        self.eeg_data = torch.tensor(eeg_data, dtype=torch.float32)
        self.panss_scores = torch.tensor(panss_scores.values, dtype=torch.long)

    def __len__(self):
        return len(self.eeg_data)

    def __getitem__(self, idx):
        return self.eeg_data[idx], self.panss_scores[idx]

def load_eeg_epochs(file_path):
    try:
        epochs = mne.read_epochs_eeglab(file_path)
        return epochs
    except Exception as e:
        print(f"Invalid EEGLAB file: {file_path}. Error: {e}")
        return None

def segment_data(epochs, tmin=-0.150, tmax=0.350, baseline=None):
    epochs = epochs.crop(tmin=tmin, tmax=tmax).apply_baseline(baseline)
    return epochs

def compute_average(epochs):
    averaged_data = epochs.get_data().mean(axis=1)
    return averaged_data

def load_panss_scores(file_path):
    df = pd.read_excel(file_path)
    return df

def extract_features(data, fs):
    _, _, Zxx = stft(data, fs=fs, nperseg=128)
    freq_features = np.abs(Zxx)  # Shape: (num_channels, freq_bins, time_points)
    power = np.mean(np.abs(Zxx)**2, axis=-1)  # Shape: (num_channels, freq_bins)
    mean = np.mean(data, axis=-1, keepdims=True)  # Shape: (num_channels, 1)
    std = np.std(data, axis=-1, keepdims=True)  # Shape: (num_channels, 1)
    
    # Reshape power to match freq_features dimensions
    power = np.expand_dims(power, axis=-1)  # Shape: (num_channels, freq_bins, 1)
    
    # Repeat mean and std to match freq_features dimensions
    mean = np.repeat(mean, freq_features.shape[1], axis=-1)  # Shape: (num_channels, freq_bins)
    mean = np.expand_dims(mean, axis=-1)  # Shape: (num_channels, freq_bins, 1)
    std = np.repeat(std, freq_features.shape[1], axis=-1)  # Shape: (num_channels, freq_bins)
    std = np.expand_dims(std, axis=-1)  # Shape: (num_channels, freq_bins, 1)
    
    # Concatenate along the last axis
    features = np.concatenate([freq_features, power, mean, std], axis=-1)  # Shape: (num_channels, freq_bins, num_features)
    return features

def process_files(folder_path, excel_file, panss_column_name, target_epochs=50):
    all_features = []
    matched_scores = []
    
    df = load_panss_scores(excel_file)
    print(f"Total rows in PANSS scores file: {len(df)}")
    
    for index, row in df.iterrows():
        mr_id = row['MR ID']
        panss_score = row[panss_column_name]
        matching_files = [f for f in os.listdir(folder_path) if f.startswith(f'AB3_{mr_id}') and f.endswith('.set')]
        if matching_files:
            file_name = matching_files[0]
            file_path = os.path.join(folder_path, file_name)
            print(f"Processing file: {file_path}")
            epochs = load_eeg_epochs(file_path)
            if epochs is None:
                print(f"Failed to load epochs from {file_path}")
                continue
            epochs = segment_data(epochs, tmin=-0.150, tmax=0.350, baseline=None)
            if len(epochs) < target_epochs:
                print(f"File {file_path} has fewer than {target_epochs} epochs. Skipping.")
                continue
            epochs = epochs[:target_epochs]
            averaged_data = compute_average(epochs)
            print(f"Averaged data shape: {averaged_data.shape}")
            features = extract_features(averaged_data, epochs.info['sfreq'])
            print(f"Feature data shape: {features.shape}")
            all_features.append(features)
            matched_scores.append(panss_score)
        else:
            print(f"No matching .set file found for MR ID: {mr_id}")
    
    if all_features:
        all_features = np.array(all_features)
        if len(all_features) != len(matched_scores):
            print(f"Inconsistent lengths: {len(all_features)} EEG data points, {len(matched_scores)} PANSS scores")
            return None, None
        matched_scores = pd.Series(matched_scores)
        return all_features, matched_scores
    else:
        print("No matching files found.")
        return None, None

def normalize_features(features):
    scaler = StandardScaler()
    num_samples, num_epochs, num_channels, num_features = features.shape
    features = features.reshape(-1, num_features)
    features = scaler.fit_transform(features)
    features = features.reshape(num_samples, num_epochs, num_channels, num_features)
    return features

class CNNLSTMModel(nn.Module):
    def __init__(self, num_channels, sequence_length, hidden_size, num_layers, dropout_rate=0.5):
        super(CNNLSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=64, kernel_size=(3, 3), padding=1)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=(3, 3), padding=1)
        self.pool = nn.MaxPool2d((2, 2), stride=(2, 2), padding=0)
        self.dropout = nn.Dropout(dropout_rate)
        
        self.conv_output_size = self._get_conv_output(num_channels, sequence_length)
        
        self.lstm = nn.LSTM(self.conv_output_size, hidden_size, num_layers, batch_first=True)
        self.fc1 = nn.Linear(hidden_size, 64)
        self.fc2 = nn.Linear(64, 4)  # Output layer for 4 classes
    
    def _get_conv_output(self, num_channels, sequence_length):
        x = torch.zeros(1, 1, num_channels, sequence_length)
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        return x.numel()
    
    def forward(self, x):
        batch_size, num_epochs, num_channels, sequence_length = x.size()
        x = x.view(batch_size * num_epochs, 1, num_channels, sequence_length)
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.dropout(x)
        x = x.view(batch_size, num_epochs, -1)
        h0 = torch.zeros(self.num_layers, batch_size, self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, batch_size, self.hidden_size).to(x.device)
        x, _ = self.lstm(x, (h0, c0))
        x = F.relu(self.fc1(x[:, -1, :]))
        x = self.fc2(x)
        return x

def get_severity_level(panss_score):
    if 16 <= panss_score <= 58:
        return 0  # Mildly ill
    elif 58 < panss_score <= 75:
        return 1  # Moderately ill
    elif 75 < panss_score <= 95:
        return 2  # Markedly ill
    elif 95 < panss_score <= 116:
        return 3  # Severely ill
    else:
        return -1  # Out of range

# Evaluation function
def evaluate_classification_metrics(predictions, actuals):
    accuracy = accuracy_score(actuals, predictions)
    f1 = f1_score(actuals, predictions, average='weighted')
    cm = confusion_matrix(actuals, predictions)
    print(f'Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}')
    print('Confusion Matrix:')
    print(cm)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
    
    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    actuals = np.array(actuals)
    predictions = np.array(predictions)
    for i in range(4):
        fpr[i], tpr[i], _ = roc_curve((actuals == i).astype(int), (predictions == i).astype(int))
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # Plot all ROC curves
    plt.figure()
    colors = ['aqua', 'darkorange', 'cornflowerblue', 'green']
    for i, color in zip(range(4), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=2,
                 label=f'ROC curve of class {i} (area = {roc_auc[i]:0.2f})')
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

# Training and Evaluation
def train_model(model, train_loader, criterion, optimizer, num_epochs):
    model.train()
    for epoch in range(num_epochs):
        running_loss = 0.0
        for i, (data, labels) in enumerate(train_loader):
            data = data.to(device)
            labels = labels.to(device).long()
            optimizer.zero_grad()
            outputs = model(data)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        epoch_loss = running_loss / len(train_loader)
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}')

def evaluate_model(model, val_loader):
    model.eval()
    predictions = []
    actuals = []
    with torch.no_grad():
        for data, labels in val_loader:
            data = data.to(device)
            labels = labels.to(device).long()
            outputs = model(data)
            _, predicted = torch.max(outputs, 1)
            predictions.extend(predicted.cpu().numpy())
            actuals.extend(labels.cpu().numpy())
    return predictions, actuals

# define path
folder_path = "G:\\AB3_MMN"
excel_file = "G:\\PANN-Patients id.xlsx"
panss_column_name = 'PANSS total'
target_epochs = 50

eeg_data, panss_scores = process_files(folder_path, excel_file, panss_column_name, target_epochs)
if eeg_data is not None and panss_scores is not None:
    eeg_data = normalize_features(eeg_data)

    # Convert PANSS scores to severity levels
    severity_levels = panss_scores.apply(get_severity_level)
    valid_indices = severity_levels != -1
    eeg_data = eeg_data[valid_indices]
    severity_levels = severity_levels[valid_indices]

    # Split the data into training and validation sets
    train_data, val_data, train_labels, val_labels = train_test_split(eeg_data, severity_levels, test_size=0.2, random_state=42)

    train_dataset = EEGDataset(train_data, train_labels)
    val_dataset = EEGDataset(val_data, val_labels)

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

    batch_size = 32
    num_epochs = 10000
    learning_rate = 0.001
    
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    
    model = CNNLSTMModel(num_channels=eeg_data.shape[2], sequence_length=eeg_data.shape[3], hidden_size=128, num_layers=2).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-5)  # Added L2 regularization

    # Train the model
    train_model(model, train_loader, criterion, optimizer, num_epochs)

    # Evaluate the model
    predictions, actuals = evaluate_model(model, val_loader)

    # Evaluate classification metrics
    evaluate_classification_metrics(predictions, actuals)
else:
    print("No valid EEG data and PANSS scores to train the model.")
```
