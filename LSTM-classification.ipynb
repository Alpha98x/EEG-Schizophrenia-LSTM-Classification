```python
import os
import numpy as np
import mne
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Subset
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import random
from imblearn.over_sampling import SMOTE

# Custom dataset class for EEG data
class EEGDataset(Dataset):
    def __init__(self, data_dir, window_size, sample_rate):
        self.data_dir = data_dir
        self.window_size = window_size
        self.sample_rate = sample_rate
        self.files = [f for f in os.listdir(data_dir) if f.endswith('.set') or f.endswith('.fdt')]
        self.data, self.labels, self.input_size = self.preprocess_data()
        
    def preprocess_data(self):
        data_list = []
        labels_list = []
        segment_length = self.window_size * self.sample_rate
        max_length = 3 * 60 * self.sample_rate  # First two minutes of data

        max_channels = 0
        
        for file in self.files:
            file_path = os.path.join(self.data_dir, file)
            
            try:
                raw = mne.io.read_raw_eeglab(file_path, preload=True)
                data = raw.get_data()
                
                # Handle missing channels by filling with zeros
                data = np.nan_to_num(data)
                
                # Truncate to first two minutes
                data = data[:, :max_length]
                
                # Update max_channels to ensure all segments have the same number of channels
                if data.shape[0] > max_channels:
                    max_channels = data.shape[0]
                
                # Normalization (z-score)
                data = (data - np.mean(data, axis=1, keepdims=True)) / np.std(data, axis=1, keepdims=True)
                
                # Segmentation
                num_windows = data.shape[1] // segment_length
                for i in range(num_windows):
                    start = i * segment_length
                    end = (i + 1) * segment_length
                    segment = data[:, start:end]
                    
                    # Check if the segment length matches the expected length
                    if segment.shape[1] != segment_length:
                        print(f"Skipping segment with shape: {segment.shape}")
                        continue
                    
                    data_list.append(segment)
                    
                    # Assign labels based on file naming convention
                    label = 0 if file.startswith('AB1') or file.startswith('AB2') else 1
                    labels_list.append(label)
            
            except Exception as e:
                print(f"Error reading {file_path}: {e}")
        
        # Ensure all segments are consistent in shape by padding if necessary
        data_array = np.zeros((len(data_list), max_channels, segment_length))
        for i, segment in enumerate(data_list):
            if segment.shape[0] < (max_channels):
                padding = np.zeros((max_channels - segment.shape[0], segment_length))
                segment = np.concatenate((segment, padding), axis=0)
            data_array[i] = segment

        labels_array = np.array(labels_list)

        # Debug: print final data shapes
        print(f"Final data shape: {data_array.shape}")
        print(f"Final labels shape: {labels_array.shape}")

        return data_array, labels_array, max_channels
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        return torch.tensor(self.data[idx], dtype=torch.float32), torch.tensor(self.labels[idx], dtype=torch.long)

# Define the RNN model
class EEGClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(EEGClassifier, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, num_classes)
    
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

# Training function
def train_model(model, train_loader, criterion, optimizer, num_epochs):
    model.train()
    epoch_losses = []
    for epoch in range(num_epochs):
        running_loss = 0.0
        for i, (data, labels) in enumerate(train_loader):
            data = data.permute(0, 2, 1)
            data, labels = data.to(device), labels.to(device)
            
            outputs = model(data)
            loss = criterion(outputs, labels)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
        
        epoch_loss = running_loss / len(train_loader)
        epoch_losses.append(epoch_loss)
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}')
    
    return epoch_losses

# Evaluation function
def evaluate_model(model, val_loader):
    model.eval()
    all_probs = []
    all_labels = []
    with torch.no_grad():
        for data, labels in val_loader:
            data = data.permute(0, 2, 1)
            data, labels = data.to(device), labels.to(device)
            outputs = model(data)
            probs = torch.softmax(outputs, dim=1)[:, 1]  # Probabilities for the positive class
            all_probs.extend(probs.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    return all_probs, all_labels

# Set the random seed for reproducibility
random_seed = 42
torch.manual_seed(random_seed)
np.random.seed(random_seed)
random.seed(random_seed)

# Hyperparameters
data_dir = 'G:\\ABP_PreprocessedEEG'  # Updated data directory
window_size = 1  # 1-second window
sample_rate = 250  # 250 Hz
hidden_size = 128
num_layers = 2
num_classes = 2  # Healthy vs Patient
batch_size = 32
num_epochs = 60
learning_rate = 0.001
n_splits = 5  # Number of folds for cross-validation

# Dataset
dataset = EEGDataset(data_dir, window_size, sample_rate)

# Cross-validation setup
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_seed)

all_fold_metrics = []
all_fprs = []
all_tprs = []
all_auc_scores = []
total_cm = np.zeros((2, 2))  # Initialize confusion matrix for averaging

for fold, (train_idx, val_idx) in enumerate(skf.split(dataset.data, dataset.labels)):
    print(f"Starting fold {fold + 1}/{n_splits}...")
    
    # Subset the data for this fold
    train_data = dataset.data[train_idx]
    train_labels = dataset.labels[train_idx]
    val_data = dataset.data[val_idx]
    val_labels = dataset.labels[val_idx]

    # Reshape data to 2D for SMOTE, then reshape back to original shape after resampling
    n_samples, n_channels, n_features = train_data.shape
    smote = SMOTE(random_state=random_seed)
    train_data_resampled, train_labels_resampled = smote.fit_resample(
        train_data.reshape(n_samples, -1), train_labels
    )
    train_data_resampled = train_data_resampled.reshape(-1, n_channels, n_features)
    
    # Create new datasets with the resampled data
    train_subset = [(torch.tensor(train_data_resampled[i], dtype=torch.float32), torch.tensor(train_labels_resampled[i], dtype=torch.long)) for i in range(len(train_labels_resampled))]
    val_subset = [(torch.tensor(val_data[i], dtype=torch.float32), torch.tensor(val_labels[i], dtype=torch.long)) for i in range(len(val_labels))]

    # Create DataLoaders
    train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_subset, batch_size=batch_size, shuffle=False)
    
    # Model, loss function, and optimizer
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = EEGClassifier(dataset.input_size, hidden_size, num_layers, num_classes).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    
    # Train the model
    epoch_losses = train_model(model, train_loader, criterion, optimizer, num_epochs)
    
    # Evaluate the model
    probs, labels = evaluate_model(model, val_loader)
    
    # Calculate metrics for this fold
    preds = np.array(probs) > 0.5
    accuracy = accuracy_score(labels, preds)
    precision = precision_score(labels, preds)
    recall = recall_score(labels, preds)
    f1 = f1_score(labels, preds)
    
    fold_metrics = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }
    
    all_fold_metrics.append(fold_metrics)
    
    print(f"Fold {fold + 1} - Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}")
    
    # Plot the ROC curve for this fold
    fpr, tpr, _ = roc_curve(labels, probs)
    roc_auc = auc(fpr, tpr)
    all_fprs.append(fpr)
    all_tprs.append(tpr)
    all_auc_scores.append(roc_auc)
    
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic - Fold {fold + 1}')
    plt.legend(loc="lower right")
    plt.show()
    
    # Plot confusion matrix for this fold
    cm = confusion_matrix(labels, preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix - Fold {fold + 1}')
    plt.show()

    # Add the confusion matrix to the total for averaging
    total_cm += cm

# Average the confusion matrices
average_cm = total_cm / n_splits

# Plot average ROC curve across all folds
mean_fpr = np.linspace(0, 1, 100)
mean_tpr = np.mean([np.interp(mean_fpr, fpr, tpr) for fpr, tpr in zip(all_fprs, all_tprs)], axis=0)
mean_auc = auc(mean_fpr, mean_tpr)

plt.figure()
plt.plot(mean_fpr, mean_tpr, color='darkorange', lw=2, label='Mean ROC curve (area = %0.2f)' % mean_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Average Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()

# Plot the training loss over epochs
plt.figure()
plt.plot(range(1, num_epochs + 1), epoch_losses, marker='o')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.show()

# Plot the average confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=average_cm, display_labels=[0, 1])
disp.plot(cmap=plt.cm.Blues)
plt.title('Average Confusion Matrix')
plt.show()

# Print average metrics across all folds
avg_accuracy = np.mean([m['accuracy'] for m in all_fold_metrics])
avg_precision = np.mean([m['precision'] for m in all_fold_metrics])
avg_recall = np.mean([m['recall'] for m in all_fold_metrics])
avg_f1 = np.mean([m['f1'] for m in all_fold_metrics])
avg_auc = np.mean(all_auc_scores)

print(f"Average Accuracy: {avg_accuracy:.4f}")
print(f"Average Precision: {avg_precision:.4f}")
print(f"Average Recall: {avg_recall:.4f}")
print(f"Average F1 Score: {avg_f1:.4f}")
print(f"Average AUC: {avg_auc:.4f}")
```
